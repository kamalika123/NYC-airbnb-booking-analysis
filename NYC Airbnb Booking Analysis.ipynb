```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
airbnb = pd.read_csv(r"AB_NYC_2019.csv")
airbnb
```

```python
airbnb.drop(['latitude','longitude','last_review','reviews_per_month'],axis=1,inplace=True)
```

```python
airbnb.head(10)
```

```python
#count of duplicate values
airbnb.duplicated(keep = False).sum()
```

```python


#filling the null values in name and host_name column with 'Unknown'
airbnb['name'].fillna('Unknown', inplace=True)
airbnb['host_name'].fillna('Unknown', inplace=True)


```

```python
airbnb.info()
```

```python

# Calculate the IQR (Interquartile Range)
Q1 = airbnb['price'].quantile(0.25)
Q3 = airbnb['price'].quantile(0.75)
IQR = Q3 - Q1

# Define the boundaries for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = airbnb[(airbnb['price'] < lower_bound) | (airbnb['price'] > upper_bound)]
```

```python
print(outliers)
```

```python
# Assuming 'outliers' contains the DataFrame rows identified as outliers
airbnb_cleaned = airbnb.drop(outliers.index)
```

```python
# Grouping by 'host_id' and 'host_name' to check consistency
host_id_name_consistency = airbnb.groupby(['host_id', 'host_name']).size().reset_index(name='count')

# Checking for inconsistencies where a host_id has multiple different host_names
inconsistent_hosts = host_id_name_consistency[host_id_name_consistency.duplicated(subset='host_id', keep=False)]

if inconsistent_hosts.empty:
    print("No inconsistencies found between 'host_id' and 'host_name'.")
else:
    print("Inconsistencies found between 'host_id' and 'host_name':")
    print(inconsistent_hosts)
```

```python


# Check if the 'minimum_nights' column contains only positive integers
is_positive_int = all(airbnb['minimum_nights'] > 0) and all(airbnb['minimum_nights'] == airbnb['minimum_nights'].astype(int))

if is_positive_int:
    print("The 'minimum_nights' column contains only positive integers.")
else:
    print("The 'minimum_nights' column does not contain only positive integers.")


```

# Information about different host and areas

```python
host_areas = airbnb.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
host_areas.sort_values(by='calculated_host_listings_count',ascending=False).head(10)
```

# Room Prices according to the area

# Room Types and Neighbourhoods


```python
plt.figure(figsize=(10, 6))
sns.countplot(x='room_type', data=airbnb, hue='neighbourhood_group')
plt.title('Room Types by Neighbourhood Group')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.legend(title='Neighbourhood Group')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(x='room_type', data=airbnb)
plt.title('Room Types Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')

plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='neighbourhood_group', y='number_of_reviews', data=airbnb, estimator=sum,color='green',width=0.5)
plt.title('Number of Reviews by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Total Number of Reviews')
plt.show()
```

```python
review_number = airbnb.groupby('neighbourhood_group')['number_of_reviews'].max().reset_index()
review_number.sort_values(by='number_of_reviews',ascending=False).head(10).reset_index(drop=True)
```

# Availability Throughout the Year

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=airbnb.index, y='availability_365', data=airbnb.sort_values('availability_365'))
plt.title('Availability Throughout the Year')
plt.xlabel('Listing Index (Sorted by Availability)')
plt.ylabel('Availability (in days)')
plt.show()
```

```python
price_according_to_reviews= airbnb.groupby(['price'])['number_of_reviews'].max().reset_index()
price_according_to_reviews.sort_values(by='number_of_reviews',ascending=False).head(10).reset_index(drop=True)
```

```python
price_list = price_according_to_reviews['price']
review = price_according_to_reviews['number_of_reviews']
fig=plt.figure(figsize=(8,5))

plt.scatter(price_list,review)
plt.xlabel('Price')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews VS Price')
plt.show()
```

```python
traffic_areas = airbnb.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_areas.sort_values(by='minimum_nights',ascending=False).head(10).reset_index(drop=True)
```

```python

plt.figure(figsize=(10, 5))
sns.barplot(x=traffic_areas['neighbourhood_group'], y=traffic_areas['minimum_nights'], data=airbnb, estimator=sum,color='purple',width=0.8)

plt.xlabel('Areas')
plt.ylabel('Minimum night Counts')

plt.show()
```
